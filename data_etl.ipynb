```python
csv_file3 = "Resources/stadium.csv"
stadium_df = pd.read_csv(csv_file3)
stadium_df.head()
```

```python
csv_file3 = "Resources/team.csv"
team_df = pd.read_csv(csv_file3)
team_df.head()
```

```python
csv_file3 = "Resources/game.csv"
game_df = pd.read_csv(csv_file3)
game_df.head()
```

```python
team_cols = ["Team Code", "Name"]
team_transformed= team_df[team_cols].copy()

team_transformed = team_transformed.rename(columns={"Team Code": "team_id",
                                                          "Name": "team_name"})

team_transformed.drop_duplicates("team_id", inplace=True)
team_transformed.set_index("team_id", inplace=True)

team_transformed.head()
```

```python
stadium_cols = ["Stadium Code", "Name", "Capacity", 'City']
staium_transformed = stadium_df[stadium_cols].copy()


staium_transformed = staium_transformed.rename(columns={"Stadium Code": "stadium_id",
                                                         "Name": "stadium_name",
                                                         "Capacity": "capacity",
                                                         "City": "city"})

staium_transformed.set_index("stadium_id", inplace=True)

staium_transformed.head()
```

```python
game_cols = ["Visit Team Code", "Home Team Code", "Stadium Code"]
game_transformed = game_df[game_cols].copy()

game_transformed = game_transformed.rename(columns={"Visit Team Code": "visit_id",
                                                         "Home Team Code": "stadium_name",
                                                         "Stadium Code": "capacity",
                                                         "City": "city"})

game_transformed.set_index("stadium_id", inplace=True)

game_transformed.head()
```
